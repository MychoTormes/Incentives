#### The correlation between streams and musical attributes

```python
z = x.loc[:,['in_spotify_playlists']]
za = z.loc[z['in_spotify_playlists']>0]
zb = za.describe()
zc = zb.loc[['count'],:]
zc
```

```python
zd = x.loc[:,['in_spotify_charts']]
zf = zd.loc[zd['in_spotify_charts']>0]
zg = zf.describe()
zh = zg.loc[['count'],:]
zh
```

```python
zi = x.loc[:,['in_apple_playlists']]
zj = zi.loc[zi['in_apple_playlists']>0]
zk = zj.describe()
zl = zk.loc[['count'],:]
zl
```

```python
zm = x.loc[:,['in_apple_charts']]
zn = zm.loc[zm['in_apple_charts']>0]
zo = zn.describe()
zp = zo.loc[['count'],:]
zp
```

```python
zz = pd.concat([zc,zh,zl,zp], axis=1)
zz
```

```python
zx = pd.melt(zz, value_vars=['in_spotify_playlists','in_spotify_charts','in_apple_playlists','in_apple_charts'], var_name = 'platforms', value_name = 'count').reset_index(drop=True)
zx
```

```python
plt.figure(figsize=(15,4))
plt.bar(zx['platforms'], zx['count'], color = 'green')
plt.xlabel('Music_Platforms')
plt.ylabel('Number_of_tracks')
plt.grid()
plt.show()
```

#### Most popular tracks in each Platforms

```python
d = x.loc[:,['track_name','artist(s)_name', 'key']]
d
```

```python
d.describe()
```

```python
dd = d.pivot_table(index= 'artist(s)_name', columns='key', values = 'track_name', aggfunc=lambda d: d.value_counts().to_dict()).reset_index()
de = dd.loc[:, ['A','A#','B','C#','D','D#','E','F','F#','G','G#']]
de
```

```python
df = de.describe()
dg = df.loc[['count'],:]
dh = pd.melt(dg, value_vars=['A','A#','B','C#','D','D#','E','F','F#','G','G#'], var_name='Key', value_name='Count')
dh
```

```python
plt.figure(figsize=(20,8))
plt.plot(dh['Key'], dh['Count'], color = 'Purple')
plt.xlabel('Key_Notes')
plt.ylabel('Counts')
plt.grid()
plt.show()
```

```python
f = x.loc[:,['track_name','artist(s)_name','mode']]
f
```

```python
fa = f.pivot_table(index= 'artist(s)_name', columns='mode', values = 'track_name', aggfunc=lambda d: d.value_counts().to_dict()).reset_index()
fb = fa.loc[:, ['Major','Minor']]
fb
```

```python
fc = fb.describe()
fd = fc.loc[['count'],:]
ff = pd.melt(fd, value_vars=['Major','Minor'], var_name='Mode', value_name='Count')
ff
```

```python
plt.figure(figsize=(10,8))
plt.bar(ff['Mode'], ff['Count'], color = 'brown')
plt.xlabel('Modes')
plt.ylabel('Counts')
plt.grid()
plt.show()
```

#### Pattern among track with same key or modes

```python
r = x.loc[:,['artist(s)_name','in_spotify_playlists','in_spotify_charts','in_apple_playlists','in_apple_charts']]
r
```

```python
ra = pd.melt(r, id_vars=['artist(s)_name'], value_vars=['in_spotify_playlists','in_spotify_charts','in_apple_playlists','in_apple_charts'], var_name='platforms', value_name='value').sort_values(by='artist(s)_name').reset_index(drop=True)
rb = ra.pivot_table(index=['platforms'], columns='artist(s)_name', values='value').reset_index()
rb
```

```python
rc = (rb != 0).sum()
rd = pd.DataFrame(rc, columns=['platform appearance']).reset_index()
re = rd.drop(0)
re
```

```python
rea = pd.DataFrame((re == 4).sum(), columns=['Four'])
reb = pd.DataFrame((re == 3).sum(), columns=['Three'])
rec = pd.DataFrame((re == 2).sum(), columns=['Two'])
red = pd.DataFrame((re == 1).sum(), columns=['One'])
ref = pd.DataFrame((re == 0).sum(), columns=['Zero'])
rff = pd.concat([rea,reb,rec,red,ref], axis=1)
rfa = rff.drop('artist(s)_name')
rf = pd.melt(rfa, value_vars=['Four','Three','Two','One','Zero'], var_name='number_of_appearances', value_name='count')
rf
```

```python
plt.figure(figsize=(10,8))
plt.bar(rf['number_of_appearances'], rf['count'], color = 'indigo')
plt.xlabel('Number_of_Platform_Appearances')
plt.ylabel('Number_of_Artists')
plt.grid()
plt.show()
```
